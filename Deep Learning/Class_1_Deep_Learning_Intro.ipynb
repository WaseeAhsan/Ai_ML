## ðŸ”¹ What is a Neural Network?

A **neural network** is a collection of connected neurons arranged in layers:

- **Input Layer**: Takes raw data (like pixel values of an image).
- **Hidden Layers**: Perform computations and extract features.
- **Output Layer**: Gives the final result (e.g., which digit?).

Each connection between neurons has a **weight**, and learning means adjusting these weights to reduce errors.

## ðŸ”¹ Why Were Neural Networks Needed?

Before neural networks, we had to **manually design features** (e.g., edge detectors, corner finders). But this:

- Was **time-consuming**
- Didn't work well for complex tasks (like handwriting or face recognition)

Neural networks **learn features automatically** by looking at examples.

A **Neural Network** is a computational model inspired by the human brain. It is composed of layers of interconnected nodes (neurons) that learn to map inputs to outputs.

## 2. ðŸ§  Structure of a Neural Network
A typical neural network includes:
- **Input Layer**: Receives data.
- **Hidden Layers**: Intermediate computations.
- **Output Layer**: Produces predictions.

Each layer has:
- **Neurons** (Nodes)  
- **Weights** & **Biases**  
- **Activation Functions**  

```

```

## 3. âš™ï¸ Activation Functions
Activation functions introduce **non-linearity** into the model.

Activation-functions-in-Neural-Networks.webp

### Common Activation Functions:

| Name        | Formula                        | Use Case |
|-------------|----------------------------------|-----------|
| **Sigmoid** | \( \sigma(x) = \frac{1}{1+e^{-x}} \) | Binary classification (output layer) |
| **Tanh**    | \( \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} \) | Hidden layers, bounded outputs |
| **ReLU**    | \( \max(0, x) \) | Most hidden layers |
| **Leaky ReLU** | \( x \, \text{if}\, x>0 \;\text{else}\, 0.01x \) | Avoids dead ReLU problems |
| **Softmax** | \( \frac{e^{x_i}}{\sum e^{x_j}} \) | Multiclass classification (output) |


### What They Do & When to Use:

- **Sigmoid**:
  - Maps input to a range between 0 and 1.
  - Good for probability outputs.
  - **Use in binary classification** at output layer.
  - Can saturate and slow training.

- **Tanh**:
  - Maps input between -1 and 1.
  - Useful when inputs are zero-centered.
  - Slightly better than Sigmoid in hidden layers.

- **ReLU (Rectified Linear Unit)**:
  - Outputs 0 for negative input, input itself otherwise.
  - Most widely used in hidden layers due to fast convergence.
  - Watch out for "dead neurons" (outputs stuck at 0).

- **Leaky ReLU**:
  - Like ReLU but allows a small gradient when input < 0.
  - **Prevents neurons from dying**.
  - Use when ReLU is causing dead neurons.

- **Softmax**:
  - Converts raw scores into probabilities that sum to 1.
  - Use in **multiclass classification problems** at output layer.

### Which to Use?
- ReLU for most hidden layers (faster convergence)  
- Sigmoid/Softmax in output layers depending on task  
- Tanh if data is centered around zero

## 6. ðŸ”„ What is Backpropagation?

Backpropagation-in-Neural-Network-1.webp

Backpropagation is the **learning algorithm** used to adjust weights using error gradients.

### Why?
To minimize the **loss function** using optimization.

### How?
- Calculate gradients using chain rule  
- Propagate from output to input  

---

## 7. ðŸ“‰ Gradient Descent & Learning Rate
Gradient Descent updates the weights using gradients to **minimize the loss**.

```python
W = W - learning_rate * dL/dW
```

### Learning Rate:
- Too high: Overshoots  
- Too low: Slow training  
- Use schedules or optimizers (e.g., Adam)  

---

## 8. ðŸŽ¯ Role of Loss Function
Loss functions measure the difference between prediction and true value.

### Common Loss Functions:
- **Binary Crossentropy**: Binary classification  
- **Categorical Crossentropy**: Multiclass classification  
- **MSE (Mean Squared Error)**: Regression  

---

## 9. ðŸ” Epochs in Training
One **epoch** = one pass through the entire dataset.

Training involves:
- Forward pass  
- Backpropagation  
- Weight updates  

More epochs = better learning (up to a point, beware overfitting!)

````python

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
````

````python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
````

````python
# Normalize the data
x_train = x_train / 255.0
x_test = x_test / 255.0
````

````python
# # Normalize the data
# x_train = x_train / 255.0
# x_test = x_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
````

###  Build the Model

#### ðŸ’¡ Explanation of Each Step:
- **Flatten(input_shape=(28, 28))**: Flattens the 28x28 pixel image into a 784-length vector so it can be input into the Dense layers.
- **Dense(128, activation='relu')**: First hidden layer with 128 neurons and ReLU activation to introduce non-linearity.
- **Dense(64, activation='relu')**: Second hidden layer with 64 neurons, again using ReLU.
- **Dense(10, activation='softmax')**: Output layer with 10 neurons (one for each digit class), using Softmax to get probabilities.

#### âš–ï¸ Model Compilation:
- **optimizer='adam'**: Adam optimizer combines momentum and RMSprop for efficient training.
- **loss='categorical_crossentropy'**: Suitable for multiclass classification with one-hot encoded labels.
- **metrics=['accuracy']**: Track accuracy metric during training and evaluation.


````python
# 1. Vanished GRadient (NN)
# 2. Exploding Gradient (NN)
````

````python
# Image -> (28_28) =784 (CNN)
# 10 class (0-9)
````

````python
# Flatten(input_shape(28,28)) | Image (Resnet )
# classifaction probability multiclass -> 10
````

````python

model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
````

````python
# epoch 10

# 1 -> Forward (loss) + Backpropagration (NEw w)
# 2-> Forward (loss) + bac ..
# ..
# 10
````

````python
# Train (80 , 20 ) | TEst data
````

````python

history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
````

````python

loss, accuracy = model.evaluate(x_test, y_test)
print("Test accuracy:", accuracy)
````

````python

plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
````
